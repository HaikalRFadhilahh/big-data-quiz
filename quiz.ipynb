```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./order.csv')
df.tail(20)
```

```python
#Jawaban Nomer 1
dg = df[['payment_type','price']]
dg.groupby(['payment_type']).mean()
```

```python
#Jawaban Nomer 2
dc = df[['order_status']].groupby(['order_status']).size()
dc
```

```python
# Jawaban Nomer 3
de = df[['product_category_name','quantity']]
de = de.loc[de['product_category_name'] == 'electronics']
de = de.groupby('product_category_name').sum()
de['total_data'] = len(df.loc[df['product_category_name'] == 'electronics'].index)
de['mean'] = de['quantity'] / de['total_data']
de
```

```python
#Jawaban Nomer 4
df[['product_category_name','quantity']].groupby('product_category_name').sum().sort_values('quantity',ascending=False).head(1)
```

```python
#Jawaban Nomer 5
dw = df[['product_category_name','product_weight_gram']]
dw.groupby('product_category_name').max().sort_values('product_weight_gram',ascending=False).head(1)
```
